# Which method to use :
## If it is basic plotting, better to use "pyplot"
## But if it is advanced, its better to use the matplotlib OO method



```python
over_50_age = heart_diesease[heart_diesease ["age"] > 50]
over_50_age
```

```python
#pyplot method
over_50_age.plot(kind ="scatter", x ="age", y ="chol", c ="target");
```

```python
# oo method
fig, ax = plt.subplots(figsize =(10,6))
over_50_age.plot(kind ="scatter", x ="age", y="chol", c="target", ax=ax);

ax.set_xlim(45,100);
```

```python
## OO method from scratch

fig , ax = plt.subplots(figsize =(10,6))
scatter = ax.scatter(x= over_50_age["age"], y = over_50_age["chol"], c=over_50_age["target"])
ax.set(title = "Heart Diesease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title='Target');
```

```python
fig, (ax0,ax1) =plt.subplots(nrows=2,ncols=1, figsize=(10,6))
scatter = ax0.scatter(x=over_50_age["age"], y=over_50_age["chol"], c=over_50_age["target"])
ax0.set(title=" Heart & Cholestrol Diesease", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50_age["chol"].mean(), linestyle="--");

#Second set

scatter = ax1.scatter(x=over_50_age["age"], y=over_50_age["thalach"], c=over_50_age["target"])
ax1.set(title =" Heart & Thalach Data", xlabel="Age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Target");
```

```python
# get the data ready
import pandas as pd
import numpy as np
heart_diesease = pd.read_csv("heart-disease (1).csv")
heart_diesease
```

```python
# Create X and Y features
# X (feature matric)

X = heart_diesease.drop("target", axis=1)

# Y (labels)

y= heart_diesease["target"]
```

```python
# Choose the right model and hyperparamater
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We will keep the default hyperparameter
clf.get_params()
```

```python
# Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

clf.fit(X_train, y_train);

y_preds = clf.predict(X_test)
y_preds
```

```python
#Evaluate the model on training data and test data

clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
# test some other metrics
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(classification_report(y_preds, y_test))
```

```python
# Improve a model
# Try difrent amount of n-estimators

np.random.seed(50)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test)* 100:2f}%" )
```

## Filling missing values with Scikit learn

```python
car_sales_missing_new = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing_new
```

```python
car_sales_missing_new.isna().sum()
```

```python
# This line of code is removing any rows from the car_sales_missing_new DataFrame where the "Price" column has missing values (NaN).
car_sales_missing_new.dropna(subset=["Price"], inplace=True)
car_sales_missing_new.isna().sum()
```

```python
X = car_sales_missing_new.drop("Price", axis=1)
y = car_sales_missing_new["Price"]
```

```python
# Filling the missing value with Scikit Learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# Fill the categorical values with " missing" and numerical values with "mean"
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy = "constant", fill_value=4)
num_imputer = SimpleImputer(strategy="mean")

# Define Columns
cat_features = ["Make", "Colour"]
door_features = ["Doors"]
num_features = ["Odometer (KM)"]

#Create an imputer (something that fills missing data)
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_imputer", door_imputer, door_features),
    ("num_imputer", num_imputer, num_features)
])

# Transform the data
filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X, columns=["Make", "Colour", "Doors", "Odometer (KM)"])
car_sales_filled
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")

transformed_X = transformer.fit_transform(car_sales_filled)
transformed_X
```

```python
# Now we have the data and numbers filled
# lets fit the model
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
