# Which method to use :
## If it is basic plotting, better to use "pyplot"
## But if it is advanced, its better to use the matplotlib OO method



```python
over_50_age = heart_diesease[heart_diesease ["age"] > 50]
over_50_age
```

```python
#pyplot method
over_50_age.plot(kind ="scatter", x ="age", y ="chol", c ="target");
```

```python
# oo method
fig, ax = plt.subplots(figsize =(10,6))
over_50_age.plot(kind ="scatter", x ="age", y="chol", c="target", ax=ax);

ax.set_xlim(45,100);
```

```python
## OO method from scratch

fig , ax = plt.subplots(figsize =(10,6))
scatter = ax.scatter(x= over_50_age["age"], y = over_50_age["chol"], c=over_50_age["target"])
ax.set(title = "Heart Diesease and Cholestrol Levels", xlabel="Age", ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title='Target');
```

```python
fig, (ax0,ax1) =plt.subplots(nrows=2,ncols=1, figsize=(10,6))
scatter = ax0.scatter(x=over_50_age["age"], y=over_50_age["chol"], c=over_50_age["target"])
ax0.set(title=" Heart & Cholestrol Diesease", xlabel="Age", ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50_age["chol"].mean(), linestyle="--");

#Second set

scatter = ax1.scatter(x=over_50_age["age"], y=over_50_age["thalach"], c=over_50_age["target"])
ax1.set(title =" Heart & Thalach Data", xlabel="Age", ylabel="Thalach")
ax1.legend(*scatter.legend_elements(), title="Target");
```

```python
# get the data ready
import pandas as pd
import numpy as np
heart_diesease = pd.read_csv("heart-disease (1).csv")
heart_diesease
```

```python
# Create X and Y features
# X (feature matric)

X = heart_diesease.drop("target", axis=1)

# Y (labels)

y= heart_diesease["target"]
```

```python
# Choose the right model and hyperparamater
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We will keep the default hyperparameter
clf.get_params()
```

```python
# Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

clf.fit(X_train, y_train);

y_preds = clf.predict(X_test)
y_preds
```

```python
#Evaluate the model on training data and test data

clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
# test some other metrics
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(classification_report(y_preds, y_test))
```

```python
# Improve a model
# Try difrent amount of n-estimators

np.random.seed(50)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test)* 100:2f}%" )
```

## Area under Receiver Opearting Characteristic Curve
* Area under Curve
* ROC

ROC curves are a comparision on model true positive rate (tpr) and false positive rate (fpr)

* True Positive - When model predicts 1 when truth is 1
* False Positive - When model predicts 1 when truth is 0
* True Negative - When model predicts 0 whrn truth is 0
* False Negative - When model predicts 0 whrn truth is 1

```python
# Create X , Y trainm test
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.metrics import roc_curve

model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)

y_prob[:15]
```

```python
y_prob_positive = y_prob[:,1]
y_prob_positive[:10]
```

```python
# Calculate the fpr, tpr, threshold 
fpr, tpr, thresholds = roc_curve(y_test, y_prob_positive)

#Check false positive rates
fpr
```

```python
# Create a function for plotting the ROC curves
import matplotlib.pyplot as plt
%matplotlib inline

def plot_roc_curve(fpr,tpr):
    """ Plots a ROC curve given the false potive rate (fpr) and true poisitive rate (tpr) of a model """

# Plot ROC curve
plt.plot(fpr, tpr , color ="Green", label="ROC")
# Plot a line with no predictive power -baseline
plt.plot([0,1],[0,1], color="orange" , label="Guessing", linestyle ="--")

#Customize the plot
plt.xlabel("False positive rate(fpr)")
plt.ylabel("True positive rate(tpr)")
plt.title("Receiver operating Characteristics Curve")
plt.legend()
plt.show()

plot_roc_curve(fpr, tpr)
```
